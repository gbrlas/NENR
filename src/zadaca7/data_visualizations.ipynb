## Zadatak 1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-8, 10, 100)
w = 2
s = [1, 0.25, 4]

for si in s:
    y = 1. / (1 + np.abs(x - w) / np.abs(si))
    plt.plot(x, y, label="s = %f" %(si))

plt.xlim(-8, 10)
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc="best")
plt.show()

```

## Zadatak 2

```python
def load_data(path):
    file = path
    data = np.loadtxt(file)
    X = data[:, [0, 1]]
    y = data[:, 2:]

    index1 = [i for i, yi in enumerate(y) if yi[0] == 1]
    index2 = [i for i, yi in enumerate(y) if yi[1] == 1]
    index3 = [i for i, yi in enumerate(y) if yi[2] == 1]
    X1 = np.array([X[i, :] for i in index1])
    X2 = np.array([X[i, :] for i in index2])
    X3 = np.array([X[i, :] for i in index3])
    
    return X1, X2, X3

def plot_clusters(path):
    X1, X2, X3 = load_data(path)

    plt.scatter(X1[:, 0], X1[:, 1], marker=',', c='r', s=40, label="100")
    plt.scatter(X2[:, 0], X2[:, 1], marker='^', c='g', s=40, label="010")
    plt.scatter(X3[:, 0], X3[:, 1], marker='o', c='b', s=40, label="001")

    plt.show()
```

```python
plot_clusters("zad7-dataset.txt")
```

```python
def plot_data(path, clusters_path, limit=True):
    param = np.loadtxt(path)
    X1, X2, X3 = load_data(clusters_path)
    w1 = []
    w2 = []
    s1 = []
    s2 = []
    
    for i in np.arange(0, len(param), 4):
        w1.append(param[i])
        s1.append(param[i+1])
        w2.append(param[i+2])
        s2.append(param[i+3])

    plt.scatter(w1, w2, c='m', s=40)
    plt.scatter(X1[:, 0], X1[:, 1], marker=',', c='r', s=40, label="100")
    plt.scatter(X2[:, 0], X2[:, 1], marker='^', c='g', s=40, label="010")
    plt.scatter(X3[:, 0], X3[:, 1], marker='o', c='b', s=40, label="001")

    if limit:
        plt.xlim(0, 1)
        plt.ylim(0, 1)

    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.legend(loc=0)
    plt.show()
```
