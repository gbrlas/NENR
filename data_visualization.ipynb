```python
import pylab
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from mpl_toolkits.mplot3d import Axes3D
```

```python
rules = [pylab.loadtxt('rules.txt')][0]
rules
```

```python
def sigmoid(x, a, b):
    return 1 / (1 + np.exp(b * (x - a)))
```

```python
X = np.arange(-4, 5, 1)
i = 1
for r in rules:
    y1 = [sigmoid(x, r[0], r[1]) for x in X]
    y2 = [sigmoid(x, r[2], r[3]) for x in X]
    plt.plot(X, y1, label="A")
    plt.plot(X, y2, label="B")
    plt.ylim(0, 1)
    plt.legend()
    plt.title("Rule " + str(i))
    plt.show()
    i += 1
```
